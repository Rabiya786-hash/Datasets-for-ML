```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Step 2: Upload CSV (if needed)
# For Colab: uncomment below if you're uploading the file manually
# from google.colab import files
# uploaded = files.upload()

# Step 3: Load Dataset
data = pd.read_csv('Salary_Data.csv')  # Make sure the file is uploaded in the same directory

# Step 4: Prepare Data
X = data[['YearsExperience']] # input
y = data['Salary']             # output

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
# Step 5: Train the Model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 6: Predict
y_pred = model.predict(X_test)

# Step 7: Plot the Results
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression - Salary Prediction')
plt.show()
```
